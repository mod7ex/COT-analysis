- **Markets**

```python
markets_df = pd.DataFrame()
markets_df['Market_Code'] = df['CFTC_Contract_Market_Code']
markets_df['Market_Name'] = df['Market_and_Exchange_Names'].str.split(' - ').str[0]

markets_df = markets_df.drop_duplicates(subset=["Market_Code"], keep="last")

markets_df.set_index('Market_Code', inplace=True)

markets_df
```

```python
import json

# Save as csv
# markets_df.to_csv(TFF_MARKETS_CSV_PATH)

markets_df.reset_index(inplace=True)

# Save as json
with open(TFF_MARKETS_JSON_PATH, "w") as f:
    json.dump(dict(zip(markets_df["Market_Name"], markets_df["Market_Code"])), f, indent=4)
```
