```python
import pandas as pd
from utils import get_df, plot_df_line_chart
from _tff import get_df_net, get_df_index, get_df_spreading, get_df_oi, Position_Types, MARKET_PARTICIPANTS
from config import PICKED_MARKET
```

```python
df = get_df()
```

```python
dealer_df = pd.DataFrame({})

PARTICIPANT = MARKET_PARTICIPANTS[2]

dealer_df['Long'] = df[f'{PARTICIPANT}_Positions_{Position_Types[0]}']
dealer_df['Short'] = df[f'{PARTICIPANT}_Positions_{Position_Types[1]}']
dealer_df['Spread'] = df[f'{PARTICIPANT}_Positions_{Position_Types[2]}']

dealer_df['Long % change'] = df[f'Change_in_{PARTICIPANT}_{Position_Types[0]}'].astype(float)
dealer_df['Short % change'] = df[f'Change_in_{PARTICIPANT}_{Position_Types[1]}'].astype(float)
dealer_df['Spread % change'] = df[f'Change_in_{PARTICIPANT}_{Position_Types[2]}'].astype(float)

dealer_df['Long % OI'] = df[f"Pct_of_OI_{PARTICIPANT}_{Position_Types[0]}"]
dealer_df['Short % OI'] = df[f"Pct_of_OI_{PARTICIPANT}_{Position_Types[1]}"]
dealer_df['Spread % OI'] = df[f"Pct_of_OI_{PARTICIPANT}_{Position_Types[2]}"]

# dealer_df['% Long'] = 100 * dealer_df['Long'] / (dealer_df['Long'] + dealer_df['Short'])
# dealer_df['% Short'] = 100 * dealer_df['Short'] / (dealer_df['Long'] + dealer_df['Short'])

dealer_df['Net'] = dealer_df['Long'] - dealer_df['Short']
dealer_df['Net % change'] = 10 * dealer_df['Net'].pct_change(periods=1)

dealer_df = dealer_df.tail(26)

dealer_df
```

```python
styled = (
    dealer_df.style
      .background_gradient(cmap="Greens", subset=['Long'], vmin=dealer_df['Long'].min(), vmax=dealer_df['Long'].max())
      .background_gradient(cmap="YlOrRd", subset=['Short'], vmin=dealer_df['Short'].min(), vmax=dealer_df['Short'].max())
      .background_gradient(cmap="Blues", subset=['Spread'], vmin=dealer_df['Spread'].min(), vmax=dealer_df['Spread'].max())

      # .background_gradient(cmap="Greens", subset=['Long % change'], vmin=dealer_df['Long % change'].min(), vmax=dealer_df['Long % change'].max())
      # .background_gradient(cmap="YlOrRd", subset=['Short % change'], vmin=dealer_df['Short % change'].min(), vmax=dealer_df['Short % change'].max())
      # .background_gradient(cmap="Blues", subset=['Spread % change'], vmin=dealer_df['Spread % change'].min(), vmax=dealer_df['Spread % change'].max())

      .background_gradient(cmap="Greens", subset=['Long % OI'], vmin=dealer_df['Long % OI'].min(), vmax=dealer_df['Long % OI'].max())
      .background_gradient(cmap="YlOrRd", subset=['Short % OI'], vmin=dealer_df['Short % OI'].min(), vmax=dealer_df['Short % OI'].max())
      .background_gradient(cmap="Blues", subset=['Spread % OI'], vmin=dealer_df['Spread % OI'].min(), vmax=dealer_df['Spread % OI'].max())

      # .background_gradient(cmap="Blues", subset=["% Long"], vmin=dealer_df["% Long"].min(), vmax=dealer_df["% Long"].max())
      # .background_gradient(cmap="Purples", subset=["% Short"], vmin=dealer_df["% Short"].min(), vmax=dealer_df["% Short"].max())

      .background_gradient(cmap="gray", subset=['Net'], vmin=dealer_df['Net'].min(), vmax=dealer_df['Net'].max())
      .background_gradient(cmap="gray", subset=['Net % change'], vmin=dealer_df['Net % change'].min(), vmax=dealer_df['Net % change'].max())
)

styled.format("{:g}")
styled.format_index("{:%Y-%m-%d}", axis=0)

print(f"{PICKED_MARKET} - {PARTICIPANT}")

styled

# numeric_cols = dealer_df.select_dtypes(include=[np.number]).columns.tolist()
```
